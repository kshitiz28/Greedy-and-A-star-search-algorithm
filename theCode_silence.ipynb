```python
def dropdown_eventhandler(change):
    print(change.new)

option_list = df.index
source = Dropdown(description="Source:", options=option_list)
source.observe(dropdown_eventhandler, names='value')
display(source)
dest = Dropdown(description="Destination:", options=option_list)
dest.observe(dropdown_eventhandler, names='value')
display(dest)
```

```python
dest1 = dest.value
print(dest1)
```

```python



source1 = source.value
final_path = [source1]
y = [None]
print(y)
while y != dest.value:
    #print(y[0])

    dfs = df.loc[df.index == source1]
    df1 = dfs.loc[:,[(dfs[col] > 0.0).all() for col in dfs.columns]]
    list = [column for column in df1.columns]
    print("the list is: ", list)
    dfs3 = dfe.loc[list]
    print(dfs3)
    dfs4 = dfs3[dest1]
    print(dfs4)
    y = dfs4.idxmin(axis = 0)
    source1 = str(y)
    print("the selected node is:",y)
    final_path.append(y)

print(source1)


final_path
```
